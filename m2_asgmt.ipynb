```python
secret = 7
guess = 3

if secret > guess:
    print('too low')
elif secret < guess:
    print('too high')
else:
    print('just right')
```

```python
small = True
green = False

if small == True and green == True:
    print('pea')
elif small == True and green == False:
    print('cherry')
elif small == False and green == True:
    print('watermelon')
else:
    print('pumpkin')
```
